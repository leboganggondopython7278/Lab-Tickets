## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
Create a Python budget tracker that asks the user for their
monthly income, allows them to input 3 expenses with names
and amounts, calculates the remaining budget, and shows a
warning if the budget is negative.
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
print("💰 PERSONAL BUDGET TRACKER")
print("="*40)

name = input("Enter your name: ")
income = float(input("Enter your monthly income: $"))

print(f"\nHello {name}! Let's track your expenses.")
print("\nEnter your 3 expenses:")
total_expenses = 0

for i in range(1, 4):
    expense_name = input(f"Expense {i} name: ")
    expense_amount = float(input(f"Amount for {expense_name}: $"))
    total_expenses += expense_amount

remaining = income - total_expenses

print("\n" + "="*40)
print(f"👤 Name:              {name}")
print(f"💵 Monthly Income:    ${income:.2f}")
print(f"💸 Total Expenses:    ${total_expenses:.2f}")
print(f"💰 Remaining Budget:  ${remaining:.2f}")

if remaining < 0:
    print("\n⚠️ WARNING: You are over budget!")
else:
    print("\n✅ Great job! You are within budget!")
```

```python
# REFINED VERSION - Your Working Budget Tracker
# print("💰 PERSONAL BUDGET TRACKER - REFINED VERSION")
print("="*40)

name = input("Enter your name: ")
income = float(input("Enter your monthly income: $"))

print(f"\nHello {name}! Let's track your expenses.")
print("\nEnter your 3 expenses:")

expenses = {}
total_expenses = 0

for i in range(1, 4):
    expense_name = input(f"Expense {i} name: ")
    expense_amount = float(input(f"Amount for {expense_name}: $"))
    expenses[expense_name] = expense_amount
    total_expenses += expense_amount

remaining = income - total_expenses

print("\n" + "="*40)
print(f"👤 Name:              {name}")
print(f"💵 Monthly Income:    ${income:.2f}")
print(f"💸 Total Expenses:    ${total_expenses:.2f}")
print(f"💰 Remaining Budget:  ${remaining:.2f}")

if remaining < 0:
    print("\n⚠️ WARNING: You are over budget!")
else:
    print("\n✅ Great job! You are within budget!")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: Show percentage of income spent"
enhancement_prompt = """
ChatGPT Prompt:
Add a feature to the budget tracker that calculates and shows
the percentage of income spent on each expense and the total
percentage of income spent.
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
# print("💰 ENHANCED PERSONAL BUDGET TRACKER")
print("="*40)

name = input("Enter your name: ")
income = float(input("Enter your monthly income: $"))

print(f"\nHello {name}! Let's track your expenses.")
print("\nEnter your 3 expenses:")

expenses = {}
total_expenses = 0

for i in range(1, 4):
    expense_name = input(f"Expense {i} name: ")
    expense_amount = float(input(f"Amount for {expense_name}: $"))
    expenses[expense_name] = expense_amount
    total_expenses += expense_amount

remaining = income - total_expenses
percentage_spent = (total_expenses / income) * 100

print("\n" + "="*40)
print(f"👤 Name:              {name}")
print(f"💵 Monthly Income:    ${income:.2f}")
print("\n📊 EXPENSE BREAKDOWN:")
for expense_name, amount in expenses.items():
    percentage = (amount / income) * 100
    print(f"  {expense_name}: ${amount:.2f} ({percentage:.1f}%)")

print(f"\n💸 Total Expenses:    ${total_expenses:.2f} ({percentage_spent:.1f}%)")
print(f"💰 Remaining Budget:  ${remaining:.2f}")

if remaining < 0:
    print("\n⚠️ WARNING: You are over budget!")
else:
    print("\n✅ Great job! You are within budget!")
```
