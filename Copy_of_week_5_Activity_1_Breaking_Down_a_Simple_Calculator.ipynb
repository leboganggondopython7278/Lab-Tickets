# The Original "Wall of Code"

Below is a basic calculator programme that does everything in one go. It displays a menu, gets input from the user, performs a calculation, and then shows the result. You can run the code if you like to see that it works.

````python
# Define the calculator functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero."
    else:
        return a / b

# Display the menu
print("Simple Calculator Menu")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

# Get the user's choice
choice = input("Enter your choice (1-4): ")

# Get the two numbers from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Perform the chosen operation
if choice == "1":
    result = add(num1, num2)
elif choice == "2":
    result = subtract(num1, num2)
elif choice == "3":
    result = multiply(num1, num2)
elif choice == "4":
    result = divide(num1, num2)
else:
    result = "Invalid choice."

# Display the result
print("Result:", result)
````

````python
# STEP 1: This function has ONE job
# Job: show the menu and get the user choice
# 'return' sends the choice BACK to the main program

def get_operation():
    print("Simple Calculator Menu")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    choice = input("Enter your choice (1-4): ")
    return choice

# Test it alone
my_choice = get_operation()
print(f"You chose: {my_choice}")
````

````python
# STEP 2: This function has ONE job
# Job: ask the user for a number and send it back
# 'prompt' is the message we want to show - like a label on a form
# float() converts text to a decimal number
# Example: "10" becomes 10.0

def get_number(prompt):
    return float(input(prompt))

# Test it alone - we call it TWICE with different messages
# Same function, different message each time!
num1 = get_number("Enter the first number: ")
num2 = get_number("Enter the second number: ")

print(f"You entered: {num1} and {num2}")
````

````python
# STEP 3: This function has ONE job
# Job: decide WHICH calculation to do based on the user choice
# Think of it like a MANAGER at a restaurant
# The manager does not cook - they just tell the RIGHT chef what to do!
# operation = the choice the user made (1, 2, 3 or 4)
# num1 and num2 = the two numbers to calculate

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

def perform_calculation(operation, num1, num2):
    if operation == "1":
        return add(num1, num2)        # sends to add chef
    elif operation == "2":
        return subtract(num1, num2)   # sends to subtract chef
    elif operation == "3":
        return multiply(num1, num2)   # sends to multiply chef
    elif operation == "4":
        return divide(num1, num2)     # sends to divide chef
    else:
        return "Invalid choice."      # no chef available!

# Test it alone
result = perform_calculation("1", 10, 5)
print(f"Result: {result}")
````

````python
# STEP 4: This function has ONE job
# Job: divide two numbers BUT protect against dividing by zero
# Think of it like a BANK
# A bank will not let you withdraw more money than you have
# Same here - we will not let you divide by zero!

# WITHOUT protection - this would CRASH the program
# WITH protection - we catch the problem and show a friendly message

def divide(a, b):
    # First we CHECK if b is zero BEFORE dividing
    # if b is zero we return a friendly error message
    if b == 0:
        return "Error: Cannot divide by zero."
    # if b is NOT zero we safely divide
    else:
        return a / b

# Test 1: Normal division
result1 = divide(10, 2)
print(f"20 divided by 4 = {result1}")

# Test 2: Divide by zero - should show error not crash!
result2 = divide(10, 0)
print(f"10 divided by 0 = {result2}")
````

# The Final Modular Code

Now, we’ll put all these functions together in our main script. Notice how each step is clear and simple. Note that our functions generally don’t print results, our menu function displays prompts, but not the user's choice; they just return values. This is good practice as it makes the functions more flexible and reusable. Run the code below to verify that it works.

````python
# COMPLETE MODULAR CALCULATOR
# All 4 jobs working together like a relay race!

# ======= THE 4 CHEFS =======
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero."
    else:
        return a / b

# ======= THE WAITER =======
def get_operation():
    print("Simple Calculator Menu")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    choice = input("Enter your choice (1-4): ")
    return choice

# ======= THE CASHIER =======
def get_number(prompt):
    return float(input(prompt))

# ======= THE MANAGER =======
def perform_calculation(operation, num1, num2):
    if operation == "1":
        return add(num1, num2)
    elif operation == "2":
        return subtract(num1, num2)
    elif operation == "3":
        return multiply(num1, num2)
    elif operation == "4":
        return divide(num1, num2)
    else:
        return "Invalid choice."

# ======= THE RELAY RACE =======
operation = get_operation()
num1 = get_number("Enter the first number: ")
num2 = get_number("Enter the second number: ")
result = perform_calculation(operation, num1, num2)
print("Result:", result)
````

````python
# =============================================
# ACTIVITY 1 - COMPLETE ANSWERS
# =============================================

# --------- REFACTORING EXPLANATION ---------
# Refactoring means reorganising code without changing what it does
# Like renovating a house - same house, better organised inside
# Our calculator still adds subtracts multiplies and divides
# But now each job has its own function - cleaner and easier to maintain!

# --------- AI COLLABORATION BEST PRACTICES ---------
# Before starting:
# Try to understand the problem yourself first
# Use specific prompts that match your skill level
# Ask for explanations not just working code

# During development:
# Break complex requests into smaller steps
# Test each function alone before combining
# Modify code slightly to check your understanding

# After implementation:
# Compare ChatGPT and Gemini responses
# Identify patterns like the manager pattern
# Think about how to extend for CELESTI project

# --------- TESTING MY UNDERSTANDING ---------
# Function Design:
# 1. Why return instead of print?
#    Return is like a chef sending food to the waiter
#    The waiter decides what to do with it next
#    Print just shows it on screen and it is gone forever

# 2. What makes a function reusable?
#    It does ONE job and works with different inputs each time
#    Like a printer that prints different documents

# 3. How do parameters make functions flexible?
#    Parameters are blank spaces filled in differently each time
#    get_number("Enter budget: ") vs get_number("Enter expense: ")
#    Same function - different message each time!

# System Architecture:
# 1. How do all functions work together?
#    Like a factory assembly line - each does ONE job
#    get_operation() → get_number() → perform_calculation() → print

# 2. What would you change to add a new operation?
#    Step 1: Add the new function
#    Step 2: Update get_operation() menu
#    Step 3: Update perform_calculation()
#    Everything else stays the same!

# 3. Which function to modify for better validation?
#    get_number() - that is where users type numbers
#    We need to catch mistakes like letters or negatives

# AI Collaboration:
# Specific constraint based prompts worked best
# AI helped me see the manager pattern in perform_calculation()
# Using two AI tools helped cross check answers

# --------- CHALLENGE 1: ADD MODULUS ---------
# Modulus = the leftover after dividing
# 10 sweets among 3 friends = 1 sweet left over
# 10 % 3 = 1

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero."
    else:
        return a / b

# NEW: Modulus function
# Returns the REMAINDER after division
def modulus(a, b):
    if b == 0:
        return "Error: Cannot divide by zero."
    return a % b

def get_operation():
    print("Simple Calculator Menu")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Modulus")  # NEW option added!
    choice = input("Enter your choice (1-5): ")
    return choice

def get_number(prompt):
    return float(input(prompt))

def perform_calculation(operation, num1, num2):
    if operation == "1":
        return add(num1, num2)
    elif operation == "2":
        return subtract(num1, num2)
    elif operation == "3":
        return multiply(num1, num2)
    elif operation == "4":
        return divide(num1, num2)
    elif operation == "5":  # NEW option added!
        return modulus(num1, num2)
    else:
        return "Invalid choice."

# --------- CHALLENGE 2: ERROR HANDLING ---------
# What happens if user types "abc" instead of a number?
# float("abc") CRASHES the program!
# We need try/except to catch this gracefully
# Like a bank form that says "please enter numbers only"

def get_number_safe(prompt):
    # Keep asking until user enters a valid number
    while True:
        try:
            # Try to convert input to float
            value = float(input(prompt))
            # If successful return the value
            return value
        except ValueError:
            # If user types letters show friendly message
            print("Please enter a valid number! Letters are not allowed.")

# --------- LOOKING AHEAD: CELESTI CONNECTION ---------
# Calculator function    →    CELESTI function
# get_operation()        →    show_menu()
# get_number()           →    get_expense_amount()
# perform_calculation()  →    process_transaction()
# divide() protection    →    calculate_budget() protection

# --------- MAIN PROGRAM ---------
operation = get_operation()
num1 = get_number_safe("Enter the first number: ")
num2 = get_number_safe("Enter the second number: ")
result = perform_calculation(operation, num1, num2)
print("Result:", result)

print("\nActivity 1 complete!")
````
