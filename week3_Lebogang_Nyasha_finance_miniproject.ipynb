## Part 2: Financial Decision Logic (15 minutes)

### Budget Limit Checker
Apply your if/else skills to real budget management:

````python
amount = float(input("Enter an expense amount: $"))

if amount >= 100:
    category = "Major Expense"
    message = "This is a major expense - plan carefully!"
elif amount >= 50:
    category = "Moderate Expense"
    message = "This is a moderate expense - consider if necessary."
elif amount >= 10:
    category = "Minor Expense"
    message = "This is a minor expense - should be fine!"
else:
    category = "Small Purchase"
    message = "This is a small purchase - go for it!"

print(f"💰 Amount: ${amount:.2f}")
print(f"📊 Category: {category}")
print(f"💡 Advice: {message}")
````

````python
print("=== Budget Limit Checker ===")

monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter expense amount: $"))

percentage = (expense / monthly_budget) * 100

if expense > monthly_budget * 0.50:
    advice = "Major purchase - think carefully!"
elif expense > monthly_budget * 0.25:
    advice = "Significant expense - consider your budget."
elif expense > monthly_budget * 0.10:
    advice = "Moderate expense - should be manageable."
else:
    advice = "Minor expense - within budget!"

print(f"\n💵 Budget: ${monthly_budget:.2f}")
print(f"💸 Expense: ${expense:.2f}")
print(f"📊 This is {percentage:.1f}% of your budget")
print(f"💡 {advice}")

````

### Savings Goal Progress
Use multiple conditions to track savings progress:

````python
print("=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))

percentage = (current_savings / goal_amount) * 100

if percentage >= 100:
    feedback = "Goal achieved! Consider a new goal!"
elif percentage >= 75:
    feedback = "Almost there! You are doing great!"
elif percentage >= 50:
    feedback = "Halfway there - keep it up!"
elif percentage >= 25:
    feedback = "Good progress, stay consistent!"
else:
    feedback = "Just getting started - you've got this!"

print(f"\n🎯 Goal: ${goal_amount:.2f}")
print(f"💰 Saved: ${current_savings:.2f}")
print(f"📊 Progress: {percentage:.1f}%")
print(f"💡 {feedback}")
````

## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
print("=== Smart Purchase Advisor ===")

available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

affordability_ratio = purchase_price / available_money

if purchase_price > available_money:
    advice = f"Save up first - you need ${purchase_price - available_money:.2f} more"
elif affordability_ratio > 0.5:
    advice = f"You can afford it but it is {affordability_ratio*100:.1f}% of your money. Consider waiting."
elif affordability_ratio > 0.2:
    advice = "This seems like a reasonable purchase for your budget!"
else:
    advice = "This is a minor expense - go for it!"

print(f"\n💵 Available Money: ${available_money:.2f}")
print(f"🛍️ Purchase Price: ${purchase_price:.2f}")
print(f"💰 Monthly Income: ${monthly_income:.2f}")
print(f"💡 Advice: {advice}")
````

````python
emergency_fund = float(input("How much is your emergency fund? $"))
recommended_emergency = monthly_income * 3

remaining_after_purchase = emergency_fund - purchase_price

if remaining_after_purchase < recommended_emergency * 0.25:
    emergency_advice = "⚠️ WARNING: This purchase will critically drain your emergency fund!"
elif remaining_after_purchase < recommended_emergency * 0.50:
    emergency_advice = "⚠️ CAUTION: This will significantly reduce your emergency fund."
else:
    emergency_advice = "✅ Your emergency fund will remain healthy after this purchase."

print(f"\n🆘 Emergency Fund: ${emergency_fund:.2f}")
print(f"📊 Recommended Fund: ${recommended_emergency:.2f}")
print(f"💡 {emergency_advice}")
````

### Sample Decision Logic to Implement

- **Can't afford**: "Save up first - you need $X more"
- **Can afford but expensive**: "You can afford it, but it's Y% of your available money. Consider waiting."
- **Reasonable purchase**: "This seems like a reasonable purchase for your budget"
- **Small purchase**: "This is a minor expense - go for it!"
